```python
median_income.rename(columns={"Zip Code":"ZIP"},inplace=True)
```

```python
median_income_2018 = median_income.sort_values(
    by=["ZIP", "2018.00"],
    ascending=True
    )[["ZIP", '2018.00']]
median_income_2018.head()
```

```python
median_income_2017 = median_income.sort_values(
    by=["ZIP", "2017.00"],
    ascending=True
    )[["ZIP", '2017.00']]
median_income_2017.head()
```

```python
median_income_2016 = median_income.sort_values(
    by=["ZIP", "2016.00"],
    ascending=True
    )[["ZIP", '2016.00']]
median_income_2016.head()
```
